```python
# cho ma tran  = [
#    [99,99,99],
#    [99,99,99],
#    [99,99,99]
#]

# gia su 0 = 1 va 1 la x
# nhận đầu vào  từ phía x và 0 luân phiên
# cho các bạn một cặp chỉ số
# nếu phía x nhập ((0,0)) thì ma trận trở thành
#[
#    [x,99,99],
#    [99,99,99],
#    [99,99,99]
#]

# nếu phía 0 nhập ((0,0))  thì yêu cầu nhập lại và nếu không thì điền vào ma trận

# nếu ai đó có 3 ô liên tiếp thì dừng trò trơi
```

```python
import json
import os

def save_game_state(matrix, turn):
    with open("game_state.json", "w") as file:
        json.dump({"matrix": matrix, "turn": turn}, file)

def load_game_state():
    try:
        with open("game_state.json", "r") as file:
            data = json.load(file)
            return data["matrix"], data["turn"]
    except (FileNotFoundError, json.JSONDecodeError):
        return [[99, 99, 99], [99, 99, 99], [99, 99, 99]], 0

def reset_game():
    if os.path.exists("game_state.json"):
        os.remove("game_state.json")
    print("Trò chơi đã được đặt lại.")

def print_matrix(matrix):
    for row in matrix:
        print(["." if cell == 99 else cell for cell in row])
    print()

def check_winner(matrix, symbol):
    for i in range(3):
        if all(matrix[i][j] == symbol for j in range(3)) or all(matrix[j][i] == symbol for j in range(3)):
            return True
    if all(matrix[i][i] == symbol for i in range(3)) or all(matrix[i][2 - i] == symbol for i in range(3)):
        return True
    return False

def get_valid_input(matrix):
    while True:
        try:
            user_input = input("Nhập vị trí (hàng cột, cách nhau bằng dấu cách, ví dụ: 1 2): ").strip()
            row, col = map(int, user_input.split())
            if 0 <= row < 3 and 0 <= col < 3 and matrix[row][col] == 99:
                return row, col
            else:
                print("Vị trí không hợp lệ hoặc đã được chọn, hãy nhập lại!")
        except ValueError:
            print("Định dạng không hợp lệ! Hãy nhập hai số nguyên cách nhau bằng dấu cách.")
        except OSError:
            print("Lỗi nhập dữ liệu. Vui lòng kiểm tra môi trường chạy chương trình.")
            return None, None

def play_game():
    while True:
        matrix, turn = load_game_state()
        players = ['X', 'O']
        
        while True:
            print_matrix(matrix)
            player = players[turn % 2]
            print(f"Lượt của {player}.")
            
            row, col = get_valid_input(matrix)
            if row is None or col is None:
                break
            
            matrix[row][col] = player
            save_game_state(matrix, turn + 1)
            
            if check_winner(matrix, player):
                print_matrix(matrix)
                print(f"{player} đã chiến thắng!")
                reset_game()
                break
            
            turn += 1
            if turn == 9:
                print_matrix(matrix)
                print("Hòa!")
                reset_game()
                break
        
        restart = input("Bạn có muốn chơi lại không? (có/không): ").strip().lower()
        if restart != "có":
            print("Cảm ơn bạn đã chơi!")
            break

if __name__ == "__main__":
    try:
        play_game()
    except OSError:
        print("Chương trình gặp lỗi nhập/xuất. Hãy kiểm tra môi trường chạy!")
```
